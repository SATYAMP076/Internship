# display all the header tags from wikipedia.org

```python
#send get requist to the webpage server to get the source code of the page
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup=BeautifulSoup(page.content)

#scrap header tag name
tag=soup.find_all(['h1','h2','h3'])

#pars header tag
tag1=[]
for i in tag:
    tag1.append(i.text.replace('\n',''))
    
#store the scraped data
wikipedia_header=pd.DataFrame({'Header':tag1},index=range(len(tag1)))
#display all the header tags from wikipedia.org
wikipedia_header
```

# IMDB’s Top rated 100 movies

```python
#send get requist to the webpage server to get the source code of the page
url="https://www.imdb.com/chart/top/"
#get the html
r=requests.get(url)

soup=BeautifulSoup(r.content,"html.parser")

#scrap movies name
scraped_movies=soup.find_all('td',class_='titleColumn')
scraped_movies


#pars movies name
movies=[]
for i in scraped_movies:
    for j in i.find_all('a'):
        movies.append(j.text.replace('\n',''))
        
#scrap movies releese year
year=soup.find_all('span',class_='secondaryInfo')
#pars movies releese year
Year=[]
for i in year:
    Year.append(i.text.strip('\n'))


# get ratings for movies
scraped_rating=soup.find_all('td',class_="ratingColumn imdbRating")

#pars ratings
ratings=[]
for rating in scraped_rating:
    rating=rating.get_text().replace('\n','')
    ratings.append(rating)


##store the scraped data
top_movies=pd.DataFrame({})
top_movies['Movies Name']=movies
top_movies['Release Year']=Year
top_movies['Ratings']=ratings
top_movies[0:100]
```

# scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp .

# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
#send get requist to the webpage server to get the source code of the page
url1="https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
#get the html
r=requests.get(url1)

soup=BeautifulSoup(r.content,"html.parser")
#scrap team name
na=soup.find_all('table',class_='table')
#pars team name
team=[]
for i in na:
    for k in i.find_all('span',class_='u-hide-phablet'):
        team.append(k.text.replace('\n',''))
        
 #scrap matches & point (tag and class same for both  )      
m=soup.find_all('tr',class_='table-body')

#pars matches
ma=[]
for i in m:
    for j in i.find('td',class_='table-body__cell u-center-text'):
        ma.append(j)
        
#pars point       
z=[]
for i in m:
    for j in i.find_all('td',class_='table-body__cell u-center-text'):
        z.append(j.text.split('\n'))    
        
p=z[1::2]
g=[]
for i in p:
    g.extend(i)

#pars rating
r=[]
for i in m:
    for j in i.find_all('td',class_='table-body__cell u-text-right rating'):
        r.append(j.text)
    

    
#scrap matches,points&rating of top team of the table(pasrs sepratly beacuse tag and class difrent from other)   
m=[soup.find('td',class_='rankings-block__banner--matches').text]
ps=[soup.find('td',class_='rankings-block__banner--points').text]
q=[soup.find('td',class_='rankings-block__banner--rating u-text-right').text.strip()]


#combine all data
mat=m+ma
pt=ps+g
ra=q+r


#store the scraped data
Men_team=pd.DataFrame({},index=range(1,21))
Men_team["TEAM"]=team
Men_team["MATCHES"]=mat
Men_team["POINTS"]=pt
Men_team["RATING"]=ra

#Top 10 ODI teams in men’s cricket along with the records for matches, points and rating
Men_team[0:10]
```

# Top 10 ODI Batsmen along with the records of their team and rating

# scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
#send get requist to the webpage server to get the source code of the page & get the html
page=requests.get('https://coreyms.com/')
soup=BeautifulSoup(page.content)



#scarap & pars Heading
heading=soup.find_all('h2',class_='entry-title')
heading1=[]
for i in heading:
    for j in i.find_all('a'):
        heading1.append(j.text.replace('\n',''))

        
        
#scarap & pars Date       
date1=soup.find_all('time',class_='entry-time')
date=[]
for i in date1:
    date.append(i.text.replace('\n',''))
    

    
    
#scarap & pars content   
content1=soup.find_all('div',class_='entry-content')
content=[]
for i in content1:
    content.append(i.text.replace('\n','')) 
    
    
    
#store scaraped data     
details_of_all_the_posts=pd.DataFrame({'HEADING':heading1,'DATE':date,'CONTENT':content},index=range(1,11))

#show post details
details_of_all_the_posts
```

# Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts

```python
#send get requist to the webpage server to get the source code of the page & get the html
page=requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
soup=BeautifulSoup(page.content)

#scarap product name
product_name_scaraped=soup.find_all('h3')
#pars product name
product_name=[]
for i in product_name_scaraped:
    product_name.append(i.text.replace('/n',''))
#remove extra comment    
product_name.remove('What choices do I have as a buyer?')

#scarap price & pars
price_scaraped=soup.find_all('span',class_='discountedPriceText')
price=[]
for i in price_scaraped:
    price.append(i.text)
    
#scarap and pars image url    
image_url_scaraped=soup.find_all('img','productImgTag')
image_url=[]
for i in image_url_scaraped:
    image_url.append(i['src'])
image_url

#store scaraped data
details_of_all_the_product=pd.DataFrame({'product name':product_name,'price':price,'Image URL':image_url},index=range(1,11))

#product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts
details_of_all_the_product
```

# Write a python program to scrape mentioned details from dineout.co.in :

iii) Location
iv) Ratings
